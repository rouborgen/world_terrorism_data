```python
def region_group(x):
  if x in ['Central America & Caribbean','North America','South America']:
    return 'Americas'
  if x in ['Central Asia','East Asia','South Asia','Southeast Asia']:
    return 'Asia'
  if x in ['Eastern Europe','Western Europe']:
    return 'Europe'
  return x
```

```python
df['region_group'] = df['region_txt'].apply(region_group)
```

```python
t = df.groupby('region_group',dropna=False)['targtype1'].agg(['count','mean'])
```
